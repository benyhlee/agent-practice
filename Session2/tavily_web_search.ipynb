## LangChain TavilySearchResults

```python
from langchain_community.tools import TavilySearchResults
```

```python
web_search = TavilySearchResults(
    max_results = 1, # 검색 갯수
    search_depth = "advanced", # basic, advanced
    include_answer = False, # False(default): 검색 결과만 반환, True: 요약해서 답변
    include_raw_content = False, # True: 웹페이지 원본 포함(토큰 비용 증가 유의), False(default): 정제된 텍스트 데이터
    topic = "genearal", # news, general, finance
    days = 30, # 최근 N일로 제한. 별 의미 없음
    # include_domains = ["naver.com"]
    # exclude_domains = []
)
```

```python
# query = "2025년 8월 현재 한국의 대통령은 누구야?."
query = "2025년 하반기 이후 애플의 매출 전망과 주가 전망"

search_results = web_search.invoke(query) # [{'title':..., 'url':..., 'content':...}, {...}, ...]
```

```python
for result in search_results:
    print(result)
    print("-" * 100)
```

```python
from IPython.display import JSON
JSON(search_results).data # title, url, content, score(관련성 점수)
```

```python
for result in search_results:
    print(len(result['content']))
```

## TavilyClient

```python
# 웹페이지 컨텐츠 전체 가져오기: langchain wraper 대신 tavily api 직접 사용
from tavily import TavilyClient

# query = "2025년 하반기 미국 경제 전망"
query = "2025년 하반기 이후 애플의 매출 전망과 주가 전망"

client = TavilyClient()
res = client.search(
    query = query,
    max_results = 2,
    search_depth = "advanced",
    include_raw_content = True,
    include_images = True,
    include_image_descriptions = True
    )

print(res)
```

```python
from IPython.display import JSON
JSON(res).data
```

```python
for news in res['results']:
    print(len(news['content']))
```

## 웹 검색하고 답하는 에이전트

```python
from langchain_openai import ChatOpenAI
from tavily import TavilyClient
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

# llm 모델 선정
llm = ChatOpenAI(model="gpt-4o-mini")

# 웹 검색
client = TavilyClient()

query = "2025년 하반기 이후 삼성전자 매출 전망"

search_results = client.search(
    query, 
    max_results = 2,
    include_raw_content = True,
    search_depth="advanced"
)

contents = search_results['results']
# print(len(contents))

# 프롬프트 템플릿 작성
prompt = ChatPromptTemplate.from_template(
    """
    당신은 사용자의 질문에 친절히 답하는 조력자입니다. 주어진 웹검색 결과를 이용해 질문에 답하세요.
    핵심만 간결하게 markdown 형식으로 정리해서 답하고 반드시 출처 링크도 명시하세요.

    질문: {query}

    검색결과: {contents}
    """
)

# 체인으로 연결하여 에이전트 구성
agent = prompt | llm | StrOutputParser()

# 웹 검색 기반 답변 구하기
response = agent.invoke({"query": query, "contents": contents})
print(response)
```

2025년 하반기 이후 삼성전자의 매출 전망에 대해 여러 전문가들은 반등 가능성을 보고 있습니다. 특히, 메모리 가격 상승과 개선된 시장 환경 덕분에 반도체 부문 실적이 회복될 것으로 기대하고 있습니다. 삼성전자는 D램 및 낸드 메모리의 경쟁력 회복에 집중할 계획이며, 반도체 시장에서의 불확실성에도 불구하고 positive한 성장세가 예상되고 있습니다.

예를 들어, KB증권은 삼성전자가 HBM 회복을 통해 하반기부터 실적 우상향을 시작할 것이라고 전망하고 있습니다. 증권가에서는 메모리 부문 성장과 파운드리 부문 적자 축소가 실적 개선에 기여할 것으로 기대하고 있습니다【1†source】【2†source】.
